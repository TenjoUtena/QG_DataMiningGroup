```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plot_x = np.linspace(-1, 6, 141)
plot_x
```

```python
plot_y = (plot_x - 2.5)**2 - 1
```

```python
plt.plot(plot_x, plot_y)
plt.show()
```

```python
def dJ(theta):
    return 2 * (theta-2.5)
```

```python
def J(theta):
    return (theta - 2.5)**2 - 1
```

```python
theta = 0.0
epsilon = 1e-8
eta = 0.1
while True:
    gradient = dJ(theta)
    last_theta = theta
    theta = theta - eta * gradient
    if(abs(J(theta) - J(last_theta)) < epsilon):
        break;
print(theta)
print(J(theta))
```

```python
def gradient_descent(initial_theta, eta, epsilon=1e-8, n_iters = 1e4):
    theta = initial_theta
    theta_history.append(initial_theta)
    i_iter = 0
    
    while i_iter < n_iters:
        gradient = dJ(theta)
        last_theta = theta
        theta = theta - eta * gradient
        theta_history.append(theta)
        
        if(abs(J(theta) - J(last_theta)) < epsilon):
            break;
        i_iter += 1
        
    if(i_iter == 1e4):
        print('超过规定循环次数\n')
        
def plot_theta_history():
    plt.plot(plot_x, J(plot_x))
    plt.plot(np.array(theta_history), J(np.array(theta_history)), color = 'g')
    plt.scatter(np.array(theta_history), J(np.array(theta_history)), color = 'r')
    plt.show()
    
```

```python
def J(theta):
    try:
        return (theta-2.5)**2 - 1
    except:
        return float('inf')
```

```python
eta = 0.1
theta_history = []
gradient_descent(0., eta)
plot_theta_history()
```
